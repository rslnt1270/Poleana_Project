```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(cv2.__version__)
```

```python
img= cv2.imread('color2.png')
print (img)
```

**Convertir formato de color BGR-RGB para trabajar Matplotlib**

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

**Convertir la imagen a escala de grises**

```python
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_grayscale, 'gray')
```

```python
U, S, Vh = np.linalg.svd(img_grayscale, full_matrices=False)
```

**DESCOMPOSICION VALORES SINGULARES (14)**

```python
k1= 14
U_k1= U[:, :k1]
S_k1 = np.diag(S[:k1])
Vh_k1 = Vh[:k1, :]

new_image1 = U_k1 @ S_k1 @ Vh_k1
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(img_grayscale, cmap='gray')

plt.subplot(1, 2, 2)
plt.title(f'Reconstrucción con {k1} valores singulares')
plt.imshow(new_image1, cmap='gray')

plt.show()
```

**PORCENTAJE DE COMPRESION**

```python
original_size = img_grayscale.size
reconstructed_size1 = U_k1.size + S[:k1].size + Vh_k1.size

compression_ratio1 = (reconstructed_size1 / original_size) * 100
print(f'Porcentaje de compresión: {compression_ratio1:.2f}%')
```
